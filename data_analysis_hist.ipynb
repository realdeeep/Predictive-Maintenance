```python
import numpy as np
from scipy.io import loadmat
import pandas as pd
```

```python
B07 = loadmat('historic_B007_3.mat')
print(B07)
```

```python
import matplotlib.pyplot as plt
de_B07 = np.array(B07['X121_DE_time'])
fe_B07 = np.array(B07['X121_FE_time'])
ba_B07 = np.array(B07['X121_BA_time'])
myarray = [i for i in range(100000)]
x = np.array(myarray)
plt.title('DE time plot')
plt.plot(x, de_B07)
```

```python
normalData = loadmat('98.mat')
print(normalData)
```

```python
de_ndata = np.array(normalData['X098_DE_time'])
fe_ndata = np.array(normalData['X098_FE_time'])
```

```python
plt.hist(de_B07, color='r', bins=100)
plt.hist(de_ndata, color='b', bins=100, alpha=0.5)
plt.show()
```

## Comparison between Fan end, Drive end and bearing data for bearing fault

```python
plt.hist(de_B07, color='b', bins=100)
plt.hist(fe_B07, color='r', bins=100)
plt.hist(ba_B07, color='g', bins=100)
plt.hist(de_ndata, color='c', bins=100, alpha=0.5)
plt.hist(fe_ndata, color='y', bins=100, alpha=0.5)
plt.show()
print(de_B07.std(), fe_B07.std(), ba_B07.std())
```

### With this we can see that
Drive end data has the most variance or in other words, it is the most significant factor in representing this fault. Let's try this with faultdata of other types as well.
